# Logistic Regression

### Data Preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../1-RawData/breast_cancer_selected.csv")
df.head()
```

```python
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#help(train_test_split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=101 )
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

### Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(solver='saga', max_iter=100)
```

```python
param_grid = {
    'C' : np.logspace(1,10, 20),
    'penalty' : ['l1', 'l2', 'elasticnet'],
    'l1_ratio' : np.linspace(1,10, 20)
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
log_grid_model = GridSearchCV(log_model, param_grid, cv=10)
log_grid_model.fit(X_train, y_train)
```

```python
log_grid_model.best_params_
```

### Performance Metrics

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
y_pred = log_grid_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm_plot = ConfusionMatrixDisplay(confusion_matrix=cm)
cm_plot.plot()
```

### Exporting the model
