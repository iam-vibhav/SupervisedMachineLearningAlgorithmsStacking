#Naming the Columns

```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

input_path = '/content/drive/My Drive/cancer/SupervisedLearningForBinaryClassification/1-RawData/wdbc.data'

col_names = [
    'id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean',
    'compactness_mean', 'concavity_mean', 'concave_points_mean', 'symmetry_mean', 'fractal_dimension_mean',
    'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se',
    'concave_points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst',
    'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst',
    'concave_points_worst', 'symmetry_worst', 'fractal_dimension_worst'
]

df = pd.read_csv(input_path, names=col_names, header=None)
df.drop(columns=['id'], inplace=True)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

output_path = '/content/drive/My Drive/cancer/SupervisedLearningForBinaryClassification/1-RawData/breast_cancer_modified.csv'

df.to_csv(output_path, index=False)



```

#Feature Selection

```python

correlation = df.corr()['diagnosis'].abs().sort_values(ascending=False)
threshold = 0.5
selected_features = correlation[correlation > threshold].index.tolist()
selected_features.remove('diagnosis')

df_selected = df[selected_features + ['diagnosis']]

output_path =  '/content/drive/My Drive/cancer/SupervisedLearningForBinaryClassification/1-RawData/breast_cancer_selected.csv'
df_selected.to_csv(output_path, index=False)
print(df_selected.isnull().sum())

```
